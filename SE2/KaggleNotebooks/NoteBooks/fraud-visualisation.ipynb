A typical organisation loses 5% of its revenue to fraud each year. It is estimated that fraud is costing the UK economy 73 billion british pounds each year.
Credit card is the most famous example.Detecting fraud is typically challenging because of these 4 characteristics :
1. Fraud causes are in a minority, sometimes only 0.001% of a companies' transactions.
2. Fraudsters will also try their best to 'blend' in and conceal their activities
3. Moreover Fraudsters change their behaviour over time , so they find new methods to avoid getting caught
4. Lastly Fraudsters tend to work together and organize their activities in a network.

Credit card companies to deal with fraud  create ruled-based systems based on location and block transactions from risky zip codes.They might also create rules to block transactions from cards used too frequently for example 'in the last 30 minutes'. Some of these rules can be highly efficient and catch a fraud, while others cannot and results in false alarm too often.

Limitations of rule-based systems:
1. Fixed threshold per rule to determine fraud
2. Limited to yes/no outcomes
3. Fail to capture interactions between features

Machine learning models  doesn't have these limitations:
1. Machine learning models adapt to data , thus they change over time
2. Uses all the data combined rather than a threshold per feature
3. Can give a score rather than a yes/no answer
4. Will typically have a better performance and can be combined with rules

```python
model=RandomForestClassifier(n_estimators=100, random_state=5)
model.fit(X_train,y_train)
predicted=model.predict(X_test)
probs=model.predict_proba(X_test)
print(roc_auc_score(y_test,probs[:,1]))
print(classification_report(y_test,predicted))
print(confusion_matrix(y_test,predicted))
```
