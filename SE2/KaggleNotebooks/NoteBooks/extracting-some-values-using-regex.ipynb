### Generating File

```python
DATA_PATH = '../input/cityofla/CityofLA/Job Bulletins'
# structured_data = pd.DataFrame(columns=sample_csv.columns, index=[i for i in range(0,683)])
structured_data = pd.DataFrame(columns=['FILE_NAME', 'JOB_CLASS_TITLE', 'JOB_CLASS_NO', 'JOB_DUTIES'], index=[i for i in range(0,683)])

error_indices = []
for index, file_name in enumerate(os.listdir(DATA_PATH)):
    with open(DATA_PATH+'/'+file_name, encoding = "ISO-8859-1") as f:
        file_data = f.read()
        try:
            structured_data.loc[index, 'FILE_NAME'] = file_name
            structured_data.loc[index, 'JOB_CLASS_TITLE'] = file_data.split('\n')[0]
            structured_data.loc[index, 'JOB_CLASS_NO'] = re.search('Class Code (\d+)', file_data).group(1)
#             structured_data.loc[index, 'JOB_DUTIES'] = re.search('DUTIES\n\s(.*\n)', file_data).group(0).split('\n\n')[1]
            structured_data.loc[index, 'JOB_DUTIES'] = re.search('DUTIES(\W+)(.*\n)', file_data).group(0).split('\n\n')[1]
        except Exception as e:
            pass
```
